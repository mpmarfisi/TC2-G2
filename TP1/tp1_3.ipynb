3. Posibles utilidades del circuito.

## Resolución

#### a)

![Circuito_a](img/ej3_a.JPG)

Planteamos el siguiente sitema de ecuaciones:

$(1) I_i = I_C+I_{R_2}$

$(2) I_C = V_i . S . C$

$(3) I_{R_2} = \frac{V_i-V_2}{R_2}$

$(4) -V_i . S . C - \frac{V_2}{R_1} = 0$

Partiendo de (4):

$(5) V_2 = - V_i . S . R_1 . C$

Reemplazando (5) en (3):

$I_{R_2} = \frac{V_i+V_i . S . R_1 . C}{R_2}$

$(6) I_{R_2} = V_i . \frac{1+S . R_1 . C}{R_2}$

Reemplazando (2) y (6) en (1)

$I_i = V_i . S . C + V_i . \frac{1+S . R_1 . C}{R_2}$

$I_i = V_i . (S . C+\frac{1+S . R_1 . C}{R_2})$

$I_i = V_i . (\frac{S . R_2 . C+1+S . R_1 . C}{R_2})$

$I_i = V_i . (\frac{1+S . (R_1+R_2) . C}{R_2})$

Finalmente:

$Z_{in}=\frac{V_i}{I_i}=\frac{R_2}{1+S . (R_1+R_2) . C}$


#### b)

![Circuito_b](img/ej3_b.JPG)

Considerando que es un OPAMP ideal, podemos decir que por sus entradas no circula corriente. Por lo tanto, no circula corriente sobre la resistencia $R_2$ y la tensión sobre la entrada inversora es igual a $V_2$. También considerando las características del OPAMP ideal, podemos decir que sobre la entrada no inversora también tenemos $V_2$.

Planteamos el siguiente sistema de ecuaciones:

$(1) I_i = I_{R_1} + I_{R_3}$

$(2) I_{R_1} = \frac{V_1}{R_1+\frac{1}{S . C_1}} = V_1 . \frac{S . C_1}{1 + S . R_1 . C_1}$

$(3) I_{R_3} = \frac{V_1-V_2}{R_3}$

$(4) \frac{V_2-V_1}{R_1} + V_2 . S . C_1 = 0$

Partiendo de (4):

$\frac{V_2}{R_1} - \frac{V_1}{R_1} + V_2 . S . C_1 = 0$

$V_2 . (\frac{1}{R_1} + S . C_1) = \frac{V_1}{R_1}$

$V_2 . (\frac{1 + S . R_1 . C_1}{R_1}) = \frac{V_1}{R_1}$

$(5) V_2 = V_1 . \frac{1}{1 + S . R_1 . C_1}$

Reemplazando (5) en (3):

$I_{R_3} = \frac{V_1}{R_3} - \frac{V_1}{R_3} . \frac{1}{1 + S . R_1 . C_1}$

$I_{R_3} = V_1 . (\frac{1}{R_3} - \frac{1}{R_3 + S . R_1 . R_3 . C_1})$

$I_{R_3} = V_1 . (\frac{R_3 + S . R_1 . R_3 . C_1 - R_3}{R_3 . (R_3 + S . R_1 . R_3 . C_1)})$

$(6) I_{R_3} = V_1 . \frac{S . R_1 . C_1}{R_3 + S . R_1 . R_3 . C_1}$

Reemplazando (2) y (6) en (1):

$I_i = V_1 . \frac{S . C_1}{1 + S . R_1 . C_1} + V_1 . \frac{S . R_1 . C_1}{R_3 + S . R_1 . R_3 . C_1}$

$I_i = V_1 . (\frac{S . C_1}{1 + S . R_1 . C_1} + \frac{R_1}{R_3} . \frac{S . C_1}{1 + S . R_1 . C_1})$

$I_i = V_1 . \frac{S . C_1}{1 + S . R_1 . C_1} . (1 + \frac{R_1}{R_3})$

$I_i = V_1 . \frac{S . C_1}{1 + S . R_1 . C_1} . \frac{R_1 + R_3}{R_3}$

Finalmente:

$Z_{in} = \frac{V_1}{I_i} = \frac{1 + S . R_1 . C_1}{S . C_1} . \frac{R_3}{R_1 + R_3} = (\frac{1}{S . C_1} + R_1) . \frac{R_3}{R_1 + R_3}$

La naturaleza de esta impedancia es capacitiva, ya que es como si estuvieran $R_1$ y $C_1$ en serie multiplicado por un factor adimencional. 

#### c)

![Circuito_c](img/ej3_c.JPG)

Planteamos el siguiente sistema de ecuaciones:

$(1) I_i = (V_2 - V_1) . S . C_1$

$(2) 0V . (\frac{1}{R_1} + \frac{1}{R_2}) - \frac{V_1}{R_1} - \frac{V_2}{R_2} = 0$

Partiendo de (2):

$(3) V_2 = - V_1 . \frac{R_2}{R_1}$

Reemplazando (3) en (1):

$I_i = (- V_1 . \frac{R_2}{R_1} - V_1) . S . C_1$

$I_i = - V_1 . (\frac{R_2}{R_1} + 1) . S . C_1$

$I_i = - V_1 . \frac{R_2 + R_1}{R_1} . S . C_1$

Finalmente:

$Z_{in} = \frac{V_1}{I_i} = - \frac{R_1}{S . (R_2 + R_1) . C_1}$

La naturaleza de esta impedancia es capacitiva ya que $\frac{R_1}{R_2+R_1}$ podría separarse como un factor adimensional que multiplica a $\frac{1}{S . C_1}$, que es la impedancia del capacitor.

#### d)

![Circuito_d](img/ej3_d.JPG)

Planteamos el siguiente sistema de ecuaciones:

$(1) I_i = \frac{V_1 - V_2}{R_1}$

$(2) V_2 = (I_i + V_1 . gm_1) . R_2$

Reemplazando (2) en (1):

$I_i = \frac{V_1 - (I_i + V_1 . gm_1) . R_2}{R_1}$

$I_i = \frac{V_1}{R_1} - I_i . \frac{R_2}{R_1} - V_1 . gm_1 . \frac{R_2}{R_1}$

$I_i . ( 1 + \frac{R_2}{R_1}) = V_1 . (\frac{1}{R_1} - gm_1 . \frac{R_2}{R_1})$

$I_i . \frac{R_1 + R_2}{R_1} = V_1 . \frac{1}{R_1} . (1 - gm_1 . R_2)$

$I_i . (R_1 + R_2) = V_1 . (1 - gm_1 . R_2)$

Finalmente:

$Z_{in} = \frac{V_1}{I_i} = \frac{R_1 + R_2}{1 - gm_1 . R_2}$

La naturaleza de la impedancia es resistiva ya que $gm_1 . R_2$ es adimencional, por lo cual queda el numerador como resistivo puro y el denominador como algo adimensional.
