# TP1 - Ejercicio 1

## Consignas
Partiendo de las siguientes estructuras circuitales, se requiere diseñar un amplificador inversor cuya impedancia de entrada sea $Z_{in}=10K\Omega$ y su transferencia de tensión $\frac{V_2}{V_1}=-3000$.
Analizar ventajas y desventajas de cada circuito.

## Resolución

#### a)

![Circuito_a](img/ej1_a.JPG)

Como consideramos que en un OPAMP ideal la diferencia de tensión entre sus entradas es 0, entonces $V_+ = V_-$ y, por lo tanto, $V_X = 0V$. Siendo $V_X$ la tensión en la pata inversora.

Entonces:

$V_1 - V_X = R_1 . I_i$

$V_1 = R_1 . I_i$

$R_1 = \frac{V_1}{I_i}$

$R_1 = Z_{in} = 10K\Omega$

Para plantear la ganancia de tensión:

$\frac{V_X - V_1}{R_1} + \frac{V_X - V_2}{R_2} = 0$

$-\frac{V_1}{R_1} = \frac{V_2}{R_2}$

$A_V = \frac{V_2}{V_1} = -\frac{R_2}{R_1}$

$A_V = -3000 = -\frac{R_2}{R_1}$

$R_2 = 3000 . R_1 = 30M\Omega$


#### b)

![Circuito_b](img/ej1_b.JPG)

Para este circuito, la impedancia de entrada sigue siendo $R_1$.

Para la ganancia de tensión planteamos nodos de la siguiente manera.

$(1): V_A . (\frac{1}{R_1} + \frac{1}{R_2}) - \frac{V_1}{R_1} - \frac{V_B}{R_2} = 0$

$(2): V_B .(\frac{1}{R_2} + \frac{1}{R_3} + \frac{1}{R_4}) - \frac{V_A}{R_2} - \frac{V_2}{R_3} = 0$

Como $V_A = 0$ se anulan los términos correspondientes. Quedando:

$V_1 = - \frac{R_1}{R_2} . V_B$

$V_B . \frac{R_2 . R_3 + R_3 . R_4 + R_2 . R_4}{R_2 . R_3 . R_4} = \frac{V_2}{R_3}$

$V_B = V_2 . \frac{R_2 . R_4}{R_2 . R_3 + R_3 . R_4 + R_2 . R_4}$

$V_1 = - \frac{R_1}{R_2} . V_2 .\frac{R_2 . R_4}{R_2 . R_3 + R_3 . R_4 + R_2 . R_4}$

$AV = \frac{V_2}{V_1} = - \frac{R_2 . R_3 + R_3 . R_4 + R_2 . R_4}{R_1 . R_4}$

Asumiendo $R_4 = R_1 = 10K\Omega$ y $R_2 = R_3 = R$:

$3000 = \frac{R^2+R . 20K\Omega}{(10K\Omega)^2}$

$R^2+R . 20K\Omega - 3000 * (10K\Omega)^2 = 0$

Al resolver la ecuación cuadrática nos queda que: $R = 537,8K\Omega$, por lo cual:

* $R_1 = 10K\Omega$
* $R_2 = 537,8K\Omega$
* $R_3 = 537,8K\Omega$
* $R_4 = 10K\Omega$

#### c)

![Circuito_c](img/ej1_c.JPG)

Como no circula corriente por las entradas del OTA, la impedancia de entrada vuelve a ser:

$Z_{in} = \frac{V_1}{I_i} = R_1 = 10K\Omega$

Para plantear la ganancia, podemos decir que el segundo OTA está configurado como una impedancia, cuyo valor es $\frac{1}{gm_2}$.

Por lo tanto:

$V_2 = - V_1 . gm_1 . \frac{1}{gm_2}$

$A_V = \frac{V_2}{V_1} = - \frac{gm_1}{gm_2}$
