Para los siguientes dipolos activos determinar la Impedancia de excitación. Utilizar un

* $C_1 = C_2 = 1\mu F$

## Resolución

Planteando el siguiente circuito con impedancias genéricas, podemos determinar $Z_{in}$ genérica y luego reemplazar para los casos particulares.

![Circuito_generico](img/ej5_generico.jpeg)

Planteamos el siguiete sistema de ecuaciones:

$(1) \frac{V_1}{Z_5} = I_3$

$(2) I_1 . Z_1 + I_2 . Z_2 = 0$

$(3) I_2 . Z_3 + I_3 . Z_4 = 0$

Partiendo de (2):

$(4) I_2 = -\frac{I_1 . Z_1}{Z_2}$

Partiendo de (3):

$(5) I_3 = -\frac{I_2 . Z_3}{Z_4}$

Reemplazando (4) en (5):

$(6) I_3 = I_1 . \frac{Z_1 . Z_3}{Z_2 .Z_4}$

Reemplazando (6) en (1):

$\frac{V_1}{Z_5} = I_1 . \frac{Z_1 . Z_3}{Z_2 .Z_4}$

Finalmente:

$Z_{in} = \frac{V_1}{I_1} = \frac{Z_1 . Z_3 . Z_5}{Z_2 .Z_4}$

#### a)

![Circuito_a](img/ej5_a.JPG)

$Z_{in} = \frac{V_1}{I_1} = \frac{R_1 . R_2 . R_4 . S . C_1}{R_3}$

```python
# Simulación Circuito a

import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sig

R1 = 1000
R2 = 1000
R3 = 1000
R4 = 1000
C1 = 1e-06

K = R1*R2*R4*C1

start_omega = 10**-1 # r/s
stop_omega = 10**5 # r/s
cant_puntos = 100000 # puntos
omega = np.linspace( start_omega, stop_omega , cant_puntos)

sys = sig.TransferFunction([K,0],[R3])
w, mag, phase = sig.bode(sys,omega)

plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.ylim([-30,100])
plt.title("Módulo de Zin")

plt.figure() 
plt.semilogx(w, phase)  # Bode phase plot
plt.title("Fase de Zin")

plt.show()
```

#### b)

![Circuito_b](img/ej5_b.JPG)

$Z_{in} = \frac{V_1}{I_1} = \frac{R_4}{S^2 . C_1 . C_2 . R_1 . R_3}$

```python
# Simulación Circuito b

import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sig

R1 = 1000
R2 = 1000
R3 = 1000
R4 = 1000
C1 = 1e-06
C2 = 1e-06

K = (C1*C2*R1*R3)/R4

start_omega = 10**-1 # r/s
stop_omega = 10**5 # r/s
cant_puntos = 100000 # puntos
omega = np.linspace( start_omega, stop_omega , cant_puntos)

sys = sig.TransferFunction([1],[K,0,0])
w, mag, phase = sig.bode(sys,omega)

plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.ylim([-30,230])
plt.title("Módulo de Zin")

plt.figure() 
plt.semilogx(w, phase)  # Bode phase plot
plt.title("Fase de Zin")

plt.show()
```

#### c)

![Circuito_c](img/ej5_c.JPG)

Planteamos el siguiente sistema de ecuaciones:

$(1) I_1 = V_i . gm_1$

$(2) I_2 = V_o . gm_2$

$(3) V_o = I_1 . \frac{1}{S . C_1}$

Reemplazando (1) en (3):

$(4) V_o = V_i . gm_1 . \frac{1}{S . C_1}$

Reemplazando (4) en (2):

$I_2 = V_i . gm_1 . \frac{1}{S . C_1} . gm_2$

Finalmente:

$Z_{in} = \frac{V_i}{I_2} = \frac{S . C_1}{gm_1 . gm_2}$

```python
# Simulación Circuito c

import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sig

C1 = 1e-06
gm1 = 100e-03
gm2 = 100e-03

K = C1/(gm1*gm2)

start_omega = 10**-1 # r/s
stop_omega = 10**5 # r/s
cant_puntos = 100000 # puntos
omega = np.linspace(start_omega, stop_omega, cant_puntos)

sys = sig.TransferFunction([K,0],[1])
w, mag, phase = sig.bode(sys,omega)

plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.ylim([-100,30])
plt.title("Módulo de Zin")

plt.figure() 
plt.semilogx(w, phase)  # Bode phase plot
plt.title("Fase de Zin")

plt.show()
```
