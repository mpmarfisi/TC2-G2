## Consignas
Dada la siguiente red, conocida como Filtro de Variables de Estado, se pide obtener las transferencias. Analizar cualitativamente la respuesta en frecuencia de cada una indicando de qué tipo de transferencia se trata (pasa-bajo, pasa-alto, etc).
* $T_1(s) = \frac{V_1}{V_{in}}$

* $T_2(s) = \frac{V_2}{V_{in}}$

* $T_3(s) = \frac{V_3}{V_{in}}$

![Circuito](img/ej8.JPG)


## Resolución

Planteamos el siguiente sistema de ecuaciones:

$(1) \frac{V_x - V_{in}}{R_5} + \frac{V_x - V_2}{R_6} = 0$

$(2) -\frac{V_1}{R_1} - V_2 . S . C_1 = 0$

$(3) -\frac{V_2}{R_2} - V_3 . S . C_2 = 0$

$(4) \frac{V_x - V_1}{R_4} + \frac{V_x - V_3}{R_3} = 0$

ACLARACIÓN: Cada resolución (a, b y c) define distintas ecuaciones enumeradas. Estas son independientes de cada item, pero las ecuaciones 1, 2, 3 y 4 que surgen del análisis del circuito son las mismas para las distintas resoluciones.

#### a) $T_1(s) = \frac{V_1}{V_{in}}$

Partiendo de (2):

$V_2 = - V_1 . \frac{1}{S . R_1 . C_1}$

$(5) V_2 = - V_1 . \frac{1}{S . \tau_1}$

Partiendo de (3):

$V_3 = - V_2 . \frac{1}{S . R_2 . C_2}$

$(6) V_3 = - V_2 . \frac{1}{S . \tau_2}$

Reemplazando (5) en (6):

$(7) V_3 = V_1 . \frac{1}{S^2 . \tau_1 . \tau_2}$

Partiendo de (4):

$V_X . (\frac{R_3 + R_4}{R_3 . R_4}) = \frac{V_1}{R_4} + \frac{V_3}{R_3}$

Reemplazando con (7):

$V_X . (\frac{R_3 + R_4}{R_3 . R_4}) = V_1 . (\frac{1}{R_4} + \frac{1}{S^2 . R_3 . \tau_1 . \tau_2})$

$V_X . (\frac{R_3 + R_4}{R_3 . R_4}) = V_1 . (\frac{S^2 . R_3 . \tau_1 . \tau_2 + R_4}{S^2 . R_3 . R_4 .\tau_1 . \tau_2})$

$(8) V_X = V_1 . (\frac{S^2 . R_3 . \tau_1 . \tau_2 + R_4}{S^2 . (R_3 + R_4) .\tau_1 . \tau_2})$

Reemplazando (5) y (8) en (1):

$V_X . (\frac{R_5 + R_6}{R_5 . R_6}) - \frac{V_2}{R_6} = \frac{V_i}{R_5}$

$V_1 . (\frac{S^2 . R_3 . \tau_1 . \tau_2 + R_4}{S^2 . (R_3 + R_4) .\tau_1 . \tau_2}) . (\frac{R_5 + R_6}{R_5 . R_6}) + V_1 . \frac{1}{S . \tau_1} . \frac{1}{R_6} = \frac{V_i}{R_5}$

$\frac{V_1}{R_6} . (\frac{S^2 . R_3 . (R_5 + R_6) .\tau_1 . \tau_2 + R_4 . (R_5 + R_6)}{S^2 . (R_3 + R_4) .\tau_1 . \tau_2 . R_5} + \frac{1}{S . \tau_1}) = \frac{V_i}{R_5}$

$\frac{V_1}{R_6} . \frac{(S . \tau_1) . (S^2 . R_3 . (R_5 + R_6) .\tau_1 . \tau_2 + R_4 . (R_5 + R_6)) + (S . \tau_1) . (S . (R_3 + R_4). \tau_2 . R_5)}{(S . \tau_1) . (S^2 . (R_3 + R_4) .\tau_1 . \tau_2 . R_5)} = \frac{V_i}{R_5}$

$\frac{V_1}{R_6} . \frac{R_3 . (R_5 + R_6) .\tau_1 . \tau_2}{(R_3 + R_4) .\tau_1 . \tau_2} . \frac{S^2 + S . \frac{(R_3 + R_4) . \tau_2 . R_5}{R_3 . (R_5 + R_6) .\tau_1 . \tau_2} + \frac{R_4 . (R_5 + R_6)}{R_3 . (R_5 + R_6) .\tau_1 . \tau_2}}{S^2} = V_i$

Finalmente:

$T_1(s) = \frac{V_1}{V_i} = \frac{R_3 + R_4}{R_3} . \frac{R_6}{R_5 + R_6} . \frac{S^2}{S^2 + S . \frac{R_3 + R_4}{R_3} . \frac{R_5}{R_5 + R_6} . \frac{1}{\tau_1} + \frac{R_4}{R_3} . \frac{1}{\tau_1 . \tau_2}}$


$\omega_0^2 = \frac{R_4}{R_3} . \frac{1}{\tau_1 . \tau_2}$

$\omega_0 = \sqrt{\frac{R_4}{R_3} . \frac{1}{\tau_1 . \tau_2}}$

$\frac{\omega_0}{Q} = \frac{R_3 + R_4}{R_3} . \frac{R_5}{R_5 + R_6} . \frac{1}{\tau_1}$

$\frac{1}{Q} . \sqrt{\frac{R_4}{R_3} . \frac{1}{\tau_1 . \tau_2}} = \frac{R_3 + R_4}{R_3} . \frac{R_5}{R_5 + R_6} . \frac{1}{\tau_1}$

$Q = \sqrt{\frac{\frac{R_4}{R_3} . \frac{1}{\tau_1 . \tau_2}}{\frac{(R_3 + R_4)^2}{R_3^2} . \frac{R_5^2}{(R_5 + R_6)^2} . \frac{1}{\tau_1^2}}} $

$Q = \sqrt{\frac{\frac{R_4}{\tau_2}}{\frac{(R_3 + R_4)^2}{R_3} . \frac{R_5^2}{(R_5 + R_6)^2} . \frac{1}{\tau_1}}}$

$Q = \sqrt{\frac{R_4}{\tau_2} . \frac{R_3}{(R_3 + R_4)^2} . \frac{(R_5 + R_6)^2}{R_5^2} . \tau_1}$

$Q = \frac{1}{(R_3 + R_4)} . \frac{R_5 + R_6}{R_5} . \sqrt{R_3 . R_4 . \frac{\tau_1}{\tau_2}}$

$k = \frac{R_3 + R_4}{R_3} . \frac{R_6}{R_5 + R_6}$

#### b) $T_2(s) = \frac{V_2}{V_{in}}$

Partiendo de (2):

$V_1 = - V_2 . S . R_1 . C_1$

$(5) V_1 = - V_2 . S . \tau_1$

Partiendo de (3):

$V_3 = - V_2 . \frac{1}{S . R_2 . C_2}$

$(6) V_3 = - V_2 . \frac{1}{S . \tau_2}$

Reemplazando (5) y (6) en (4):

$V_X . (\frac{R_3 + R_4}{R_3 . R_4}) = \frac{V_1}{R_4} + \frac{V_3}{R_3}$

$V_X . (\frac{R_3 + R_4}{R_3 . R_4}) = - V_2 . \frac{S . \tau_1}{R_4} - V_2 . \frac{1}{S . R_3 . \tau_2}$

$V_X . (\frac{R_3 + R_4}{R_3 . R_4}) = - V_2 . (\frac{S^2 . R_3 . \tau_1 . \tau_2 + R_4}{S . R_3 . R_4 . \tau_2})$

$(7) V_X = - V_2 . \frac{S^2 . R_3 . \tau_1 . \tau_2 + R_4}{S . (R_3 + R_4) . \tau_2}$

Reemplazando (7) en (1):

$V_X . (\frac{R_5 + R_6}{R_5 . R_6}) - \frac{V_2}{R_6} = \frac{V_i}{R_5}$

$- \frac{V_2}{R_6} . (\frac{S^2 . R_3 . (R_5 + R_6) . \tau_1 . \tau_2 + R_4 . (R_5 + R_6)}{S . (R_3 + R_4) . R_5 . \tau_2}+ 1) = \frac{V_i}{R_5}$

$- \frac{V_2}{R_6} . (\frac{S^2 . R_3 . (R_5 + R_6) . \tau_1 . \tau_2 + R_4 . (R_5 + R_6) + S . (R_3 + R_4) . R_5 . \tau_2}{S . (R_3 + R_4) . R_5 . \tau_2}) = \frac{V_i}{R_5}$

Finalmente:

$T_2(s) = \frac{V_2}{V_i} = - \frac{R_3 + R_4}{R_3} . \frac{R_6}{R_5 + R_6} . \frac{1}{\tau_1} . \frac{S}{S^2 + S . \frac{R_3 + R_4}{R_3} . \frac{R_5}{R_5 + R_6} . \frac{1}{\tau_1} + \frac{R_4}{R_3} . \frac{1}{\tau_1 . \tau_2}}$

$T_2(s) = \frac{V_2}{V_i} = - \frac{R_5}{R_5} . \frac{R_3 + R_4}{R_3} . \frac{R_6}{R_5 + R_6} . \frac{1}{\tau_1} . \frac{S}{S^2 + S . \frac{R_3 + R_4}{R_3} . \frac{R_5}{R_5 + R_6} . \frac{1}{\tau_1} + \frac{R_4}{R_3} . \frac{1}{\tau_1 . \tau_2}}$

$T_2(s) = \frac{V_2}{V_i} = - \frac{R_6}{R_5} . \frac{S . \frac{R_3 + R_4}{R_3} . \frac{R_5}{R_5 + R_6} . \frac{1}{\tau_1}}{S^2 + S . \frac{R_3 + R_4}{R_3} . \frac{R_5}{R_5 + R_6} . \frac{1}{\tau_1} + \frac{R_4}{R_3} . \frac{1}{\tau_1 . \tau_2}}$


Como el denominador es el mismo $\omega_0$ y $Q$ son iguales

#### c) $T_3(s) = \frac{V_3}{V_{in}}$

Partiendo de (3):

$V_2 = - V_3 . S . R_2 . C_2$

$T_3(s) = \frac{V_3}{V_i} = \frac{R_3 + R_4}{R_3} . \frac{R_6}{R_5 + R_6} . \frac{1}{\tau_1 . \tau_2} . \frac{1}{S^2 + S . \frac{R_3 + R_4}{R_3} . \frac{R_5}{R_5 + R_6} . \frac{1}{\tau_1} + \frac{R_4}{R_3} . \frac{1}{\tau_1 . \tau_2}}$
