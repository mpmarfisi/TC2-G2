## Resolución

### Obtener la función transferencia

#### Circuito a

Planteo las ecuaciones:

(1) $\frac{(V_X - V_1)}{R_1} + \frac{V_X - V_2}{R_2} = 0$

(2) $\frac{V_X}{R_3} + (V_X-V_1) . S . C = 0$

Partiendo de la ecuación (1):

$V_X . (\frac{1}{R_1} + \frac{1}{R_2}) - \frac{V_1}{R_1} - \frac{V_2}{R_2} = 0$

$V_X . \frac{R_1 + R_2}{R_1 . R_2} - \frac{V_1}{R_1} = \frac{V_2}{R_2}$ (3)

Partiendo de la ecuación (2):

$V_X . (\frac{1}{R_3} + S . C) - V_1 . S . C = 0$

$V_X = \frac{V_1 . S . C}{\frac{1}{R_3} + S . C}$ (4)

Reemplazando (4) en (3):

$V_1 . \frac{S . C}{\frac{1}{R_3} + S . C} \frac{R_1 + R_2}{R_1 . R_2} - \frac{V_1}{R_1} = \frac{V_2}{R_2}$

$V_1 . (\frac{S . C}{\frac{1}{R_3} + S . C} \frac{R_1 + R_2}{R_1 . R_2} - \frac{1}{R_1}) = \frac{V_2}{R_2}$

$V_1 . (\frac{S . R_1^2 . C + S . R_1 . R_2 . C - S . R_1 . R_2 . C-\frac{R_1 . R_2}{R_3}}{S . R_1^2 . R_2 . C+\frac{R_1^2 . R_2}{R_3}}) = \frac{V_2}{R_2}$

$V_1 . \frac{S-\frac{R_2}{R_1 . R_3 . C}}{S + \frac{1}{R_3 . C}} = V_2$

$T(S) = \frac{V_2}{V_1} = \frac{S - \frac{R_2}{R_1 . R_3 . C}}{S + \frac{1}{R_3 . C}}$

#### Circuito b

Definimos $Z_1 = R + \frac{1}{S . C} = \frac{1 + S . R . C}{S . C}$

Definimos $Z_2 = \frac{1}{\frac{1}{R} + S . C} = \frac{R}{1 + S . R . C}$

Aplicando estas definiciones, podemos plantear el circuito equivalente como:

![Circuito_b](img/ej7_b.jpeg)

Este circuito cumple con la misma forma que el circuito a, solo que en lugar del capacitor está $Z_1$ y en lugar de $R_3$ está $Z_2$.

Planteamos el sistema de ecuaciones:

$(1) \frac{V_X - V_1}{R_A} + \frac{V_X - V_2}{R_B} = 0$

$(2) \frac{V_X - V_1}{Z_1} + \frac{V_X}{Z_2} = 0$

Partiendo de (1):

$V_X . (\frac{1}{R_A}+\frac{1}{R_B}) - \frac{V_1}{R_A} - \frac{V_2}{R_B} = 0$

$(3) \frac{V_2}{R_B} = V_X . \frac{R_A + R_B}{R_A . R_B} - \frac{V_1}{R_A}$

Partiendo de (2):

$V_X . (\frac{1}{Z_1}+\frac{1}{Z_2}) = \frac{V_1}{Z_1}$

$V_X . (\frac{S . C}{1 + S . R . C} + \frac{1 + S . R . C}{R}) = V_1 . \frac{S . C}{1 + S . R . C}$

$V_X . (\frac{S^2 . R^2 . C^2 +3 . S . R . C + 1}{R . (1 + S . R . C)}) = V_1 . \frac{S . C}{1 + S . R . C}$

$V_X = V_1 . \frac{S . R . C}{S^2 . R^2 . C^2 +3 . S . R . C + 1}$

$V_X = V_1 . \frac{1}{R^2 . C^2} . \frac{S . R . C}{S^2 +S . \frac{3}{R . C} + \frac{1}{R^2 . C^2}}$

$(4) V_X = V_1 . \frac{1}{R . C} . \frac{S}{S^2 +S . \frac{3}{R . C} + \frac{1}{R^2 . C^2}}$

Reemplazando (4) en (3):

$\frac{V_2}{R_B} = V_1 . \frac{1}{R . C} . \frac{S}{S^2 +S . \frac{3}{R . C} + \frac{1}{R^2 . C^2}} . \frac{R_A + R_B}{R_A . R_B} - \frac{V_1}{R_A}$

Definimos:

$K = \frac{1}{R . C} . \frac{R_A + R_B}{R_B}$

Entonces queda:

$\frac{V_2}{R_B} = V_1 . \frac{1}{R_A} . (K . \frac{S}{S^2 +S . \frac{3}{R . C} + \frac{1}{R^2 . C^2}} - 1)$

$\frac{V_2}{R_B} = V_1 . \frac{1}{R_A} . (\frac{-S^2 + S . (K - \frac{3}{R . C}) - \frac{1}{R^2 . C^2}}{S^2 +S . \frac{3}{R . C} + \frac{1}{R^2 . C^2}})$

Finalmente:

$T(S) = \frac{V_2}{V_1} = - \frac{R_B}{R_A} . \frac{S^2 - S . (K - \frac{3}{R . C}) + \frac{1}{R^2 . C^2}}{S^2 +S . \frac{3}{R . C} + \frac{1}{R^2 . C^2}}$


### Normalización de la función transferencia

#### Circuito a

Definimos $\omega_0 = \frac{1}{R_3 . C}$

Suponemos el caso $R_1 = R_2$

Por lo tanto, la transferencia quedaría:

$T(S) = \frac{S - \omega_0}{S + \omega_0}$

Normalizo en frecuencia: $S = S' . \omega_0$

$T(S') = \frac{S' . \omega_0 - \omega_0}{S' . \omega_0 + \omega_0}$

$T(S') = \frac{S' - 1}{S' + 1}$

#### Circuito b

Definimos $\omega_0 = \frac{1}{R . C}$

Suponemos el caso $\frac{R_A}{R_B} = 5$

Por lo tanto, la transferencia quedaría:

$T(S) = - \frac{R_B}{R_A} . \frac{S^2 - S . \frac{1}{R . C} . (\frac{R_A}{R_B} + 1 - 3) + \frac{1}{R^2 . C^2}}{S^2 + S . 3 . \frac{1}{R . C} + \frac{1}{R^2 . C^2}}$

Normalizo en frecuencia: $S = S' . \omega_0$

$T(S') = - \frac{1}{5} . \frac{S'^2 . \omega_0^2 - S' . \omega_0^2 . 3 + \omega_0^2}{S'^2 . \omega_0^2 + S' . \omega_0^2 . 3 + \omega_0^2}$

$T(S') = - \frac{1}{5} . \frac{S'^2 - S' . 3 + 1}{S'^2 + S' . 3 + 1}$

```python
# Simulación de la Transferencia Normalizada

import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sig
from splane import analyze_sys, pretty_print_bicuad_omegayq

# Señal normalizada - circuito a
num_a = np.array([1,-1])
den_a = np.array([1,1])

pretty_print_bicuad_omegayq(num_a, den_a)
transfer_f_a = sig.TransferFunction(num_a, den_a)

plt.close('all')
analyze_sys(transfer_f_a, 'Transferencia a')

```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sig
from splane import analyze_sys, pretty_print_bicuad_omegayq

# Señal normalizada - circuito b

num_b = np.array([1,-3,1])
den_b = np.array([1,3,1])

pretty_print_bicuad_omegayq(num_b, den_b)
transfer_f_b = sig.TransferFunction(num_b, den_b)

plt.close('all')
analyze_sys(transfer_f_b, 'Transferencia b')
```
