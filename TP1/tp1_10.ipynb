# TP1 - Ejercicio 10

## Consignas

Se desea implementar un integrador con pérdidas, para lo cual se propone utilizar un OTA ideal cargado con un capacitor de $C = 20pF$ y un resistor de $R = 10 K\Omega$. Calcular la transferencia $\frac{V_2}{V_1}$ del circuito.

![Circuito](img/ej10_consigna.JPG)

## Resolución

Planteamos las corrientes del cicuito:

![Corrientes](img/ej10_corrientes.jpeg)

$$
I_{1}=I_{2}+I_{3} \\
I_{1} = V_{1}*g_{m} \\
I_{2} = V_{2}*SC \\
I_{3} = V_{2}/R \\
$$
Entonces
$$
V_{1}*g_{m}=V_{2}*(SC+1/R)
$$
Asi que llegamos a que 
$$
\frac{V_{2}}{V_{1}} = \frac{gm}{C}*\frac{1}{S+1/RC}
$$
